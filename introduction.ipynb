## Now for the goodies: 

### let's create a random array and display it. 

Now lets create some data:

```python
data = np.random.random((400,600))
```

... and plot it:

```python
plt.figure()
plt.imshow(data)
```

As you can see, we're provided with a neat figure, with which we can interact. Different cells can use code that has already been processed, so lets try to get data from above and filter it with a gaussian:
(To suppress unuseful output, as above (the matplotlib thingy), add semicolons as in the example below)

```python
from scipy.ndimage.filters import gaussian_filter
data = gaussian_filter(data,5)
plt.figure()
plt.imshow(data);
```

If we now recalculate the cell that generated our plot above, we can see that the data is now gaussian filtered. The history of commands follows the numbers shown besides "In".
